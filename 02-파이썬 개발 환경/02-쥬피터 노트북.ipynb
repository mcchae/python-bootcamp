# 주피터 노트북

이전에 살펴본 pyenv 환경이 갖추어 졌다고 가정합니다.
그러면 그 위에서 다음과 같이 작업을 진행합니다.

> 모두 터미널에서 작업을 진행합니다. 이 글을 작성하는 동안 윈도우 환경에서는 pyenv가 제대로 동작하지 않았으므로 리눅스 또는 맥에서 진행하도록 합니다.

## jupyter 환경 구축

### 가상환경 구축

```sh
$ pyenv versions
  system
  2.7.13
  ...
* 3.6.2
```
와 같이 pyenv 환경이 구축되어 있습니다.

이제 다음과 같은 명령으로 `jupyter` 라는 이름의 가상환경을 만들어 봅니다.
(꼭 jupyter 일 필요는 없고 위의 versions에서 보이는 것과 같은 기존에 생성했던 것이 아닌 이름이면 됩니다.)

```sh
$ pyenv virtualenv 3.6.2 jupyter
```

그리고 다음과 같은 명령으로 해당 환경으로 들어갑니다.

```sh
$ pyenv activate jupyter
(jupyter) $
```

### jupyter 모듈 설치

위와 같이 jupyter 가상환경에서 다음과 같은 명령을 입력합니다.

```sh
(jupyter) $ pip install jupyter
...
```

성공적으로 설치되었다면,

```sh
(jupyter) $ pip list | grep jupyter
jupyter (1.0.0)
jupyter-client (5.2.2)
jupyter-console (5.2.0)
jupyter-core (4.4.0)
```

와 같이 설치된 것이 보입니다.
